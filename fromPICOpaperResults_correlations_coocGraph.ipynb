```python
import pandas as pd
import plotly.express as px
from scipy.stats import spearmanr, permutation_test  #scipy v1.17.1
```

```python
PATH_data = "../data/MTBLS28_urinaryMetabolomic/MTBLS28_187294/MTBLS28_CombinedData2_forML.csv"
```

```python
df = pd.read_csv(PATH_data, delimiter=";")
df = df.set_index("Unnamed: 0")
```

```python
corr_spearman = spearmanr(df['23.30921876_264.1215224'], df['27.98128537_372.9232556'])
print(corr_spearman.statistic)
```

```python
feature1 = df['23.30921876_264.1215224']
feature2 = df['27.98128537_372.9232556']

def statistic(feature1): # permute only `x`
    return spearmanr(feature1, feature2).statistic
    
res_exact = permutation_test((feature1,), statistic, permutation_type='pairings')
res_asymptotic = spearmanr(feature1, feature2)
res_exact.pvalue, res_asymptotic.pvalue # asymptotic pvalue is too low
```
